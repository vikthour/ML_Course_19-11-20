## Reading a file

### Read_

To read a file. For instance, read_csv reads a csv file from a specified filepath.

```python
#.head() Checks the top n-rows of a dataset

malaria = pd.read_csv(r'C:\Users\VictorAdeife\Documents\Dataset\DatasetAfricaMalaria.csv')
malaria.head(2)
```

### Describe

This function is Pandas' predefined way of providing insights into the data. It gives a general summary of each numerical feature in the dataset.

```python
malaria.describe()
```

### Info

This function is Pandas' predefined medium of providing details of each column with respect to the elements by counting the data and stating the data type in each column.

```python
malaria.info()
```

### Memory Usage

This function provides information on the memory usage per column

```python
malaria.memory_usage()
```

### Index

This function provides about the index

```python
malaria.index.memory_usage()
```

### Value Count

This gives the count of the values in each column.

```python
malaria.value_counts
```

### NUnique

This gives the number of unique elements in the column.

```python
malaria.nunique()
```

### Columns

This gives the list of the names of each feature in the dataset

```python
malaria.columns
```

### Count

This counts the number of the elements contained in each feature.

```python
malaria.count()
```

### Correlation

This function gives a mutual relationship or connection between two or more things.

```python
malaria.corr()
```

### Saving a file

This function is used to save a file in the format it is needed.
